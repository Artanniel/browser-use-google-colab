```python
import nest_asyncio
nest_asyncio.apply()

from langchain_google_genai import ChatGoogleGenerativeAI  # Substituição do import
from browser_use import Agent
from dotenv import load_dotenv
load_dotenv()

import asyncio

# Configurar o modelo Gemini
llm = ChatGoogleGenerativeAI(model="gemini-pro")  # Modelo Gemini para texto

async def main():
    agent = Agent(
        task="Compare the price of gpt-4o and DeepSeek-V3",
        llm=llm,
    )
    result = await agent.run()
    print(result)

asyncio.run(main()) # Now this should run without the error
```
